```python
from tkinter import *
from tkinter import ttk
import tkinter.messagebox
import pymysql



class ouvrage:
    def __init__(self,root):
        self.root=root
        titlespace=""
        self.root.title(102 * titlespace +"Gestion_bibliothÃ¨que")
        self.root.geometry("1200x900+0+0")
        self.root.resizable(width=False, height=False)
        
        MainFrame=Frame(self.root,bd=10,width=1200,height=900,relief= RIDGE,bg='pink')
        MainFrame.grid()

        TitleFrame=Frame(MainFrame,bd=7,width=1200,height=100,relief=RIDGE)
        TitleFrame.grid(row=0,column=0)
        TopFrame3=Frame(MainFrame,bd=5,width=1200,height=500,relief=RIDGE)
        TopFrame3.grid(row=1,column=0)
        
        LeftFrame=Frame(TopFrame3,bd=5,width=1200,height=400,padx=2,bg="pink",relief=RIDGE)
        LeftFrame.pack(side=LEFT)
        LeftFrame1=Frame(LeftFrame,bd=5,width=600,height=180,padx=2,pady=4,bg="pink",relief=RIDGE)
        LeftFrame1.pack(side=TOP,padx=0,pady=0)
        
        RightFrame1=Frame(TopFrame3,bd=5,width=100,height=400,padx=2,bg="pink",relief=RIDGE)
        RightFrame1.pack(side=RIGHT)
        RightFrame1a=Frame(RightFrame1,bd=5,width=90,height=300,padx=2,pady=2,bg="pink",relief=RIDGE)
        RightFrame1a.pack(side=TOP)
        #========================Declaration des variabels===============================================================================
        idOuvrage=IntVar()
        auteur=StringVar()
        type=StringVar()
        reference=StringVar()
        titre=StringVar()
        #=======================================================================================================
        
        def iExit():
            iExit= tkinter.messagebox.askyesno("Gestion bebliotheque","Confirm if you want to exit")
            if iExit>0:
                root.destroy()
                return
        
        def Reset():
            self.entid.delete(0,END)
            self.entauteur.delete(0,END)
            type.set(" ")
            
            self.entreference.delete(0,END)
            self.enttitre.delete(0,END)
            
            
           
        def add():
            if idOuvrage.get()=="" or auteur.get()=="" or type.get()=="" or reference.get()=="" or titre.get()=="":
                tkinter.messagebox.showerror("Enter correct details")
            else:
                    sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
                    cur=sqlcon.cursor()
                    cur.execute("insert into ouvrage values (%s, %s, %s, %s, %s)",(idOuvrage.get(),auteur.get(),type.get(),reference.get(),titre.get()))
                    
                    sqlcon.commit()
                    sqlcon.close()
                    
                    tkinter.messagebox.showinfo("Gestion bebliotheque","Record Entered succesfully")
                    
                    
        def Display():
            sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
            cur=sqlcon.cursor()
            cur.execute("select * from ouvrage")
            result=cur.fetchall()
            if len(result)!=0:
                self.id_records.delete(*self.id_records.get_children())
                for row in result:
                    self.id_records.insert('',END,values=row)
            sqlcon.commit()
            sqlcon.close()
                    
            tkinter.messagebox.showinfo("Gestion bebliotheque","Record Entered succesfully")
                    
            
            
        def OuvrageInfo(val):
            viewInfo=self.id_records.focus()
            learnerData= self.id_records.item(viewInfo)

            row=learnerData['values']
            idOuvrage.set(row[0])
            auteur.set(row[1])
            type.set(row[2])
            reference.set(row[3])
            titre.set(row[4])
             
            
        def update():
                    sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
                    cur=sqlcon.cursor()
                    cur.execute("update ouvrage set auteur=%s, type=%s, reference=%s,titre=%s where idOuvrage=%s",(auteur.get(),type.get(),reference.get(),titre.get(),idOuvrage.get()))
                    
                    sqlcon.commit()
                    sqlcon.close()
                    
                    tkinter.messagebox.showinfo("Gestion bebliotheque","Record updated successfully")
                    
                    
        def delete():
                    sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
                    cur=sqlcon.cursor()
                    cur.execute("delete from ouvrage where idOuvrage=%s",idOuvrage.get())
                    
                    sqlcon.commit()
                    Display()
                    sqlcon.close()
                    
                    tkinter.messagebox.showinfo("Gestion bebliotheque","Record deleted successfully")
                    Reset()
        
        def search():
            try:
                sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
                cur=sqlcon.cursor()
                cur.execute("select * from ouvrage where idOuvrage='%s'"%idOuvrage.get())
                row=cur.fetchone()
                    
                idOuvrage.set(row[0])
                auteur.set(row[1])
                type.set(row[2])
                reference.set(row[3])
                titre.set(row[4])
                    
                sqlcon.commit()
                    
            except:  
                tkinter.messagebox.showinfo("Data Entry Form","No such Record Find")
                Reset()
            
            
            sqlcon.close()
                    
                    
           
            
                    
        #=======================================================================================================
        self.lbltitle=Label(TitleFrame, font=('arial',40,'bold'),text='Gestion_ouvrage',bd=7)
        self.lbltitle.grid(row=0,column=0,padx=132)
        
        
        
        
        self.lblid=Label(LeftFrame1, font=('arial',12,'bold'),text='Id_Ouvrage',bd=7,bg="pink")
        self.lblid.grid(row=1,column=0,sticky=W,padx=5)
        self.entid=Entry(LeftFrame1, font=('arial',12,'bold'),bd=5,width=44, justify='left',textvariable=idOuvrage)
        self.entid.grid(row=1,column=1,sticky=W,padx=5)
        
         
        
        self.lblauteur=Label(LeftFrame1, font=('arial',12,'bold'),text='Auteur',bd=7,bg="pink")
        self.lblauteur.grid(row=2,column=0,sticky=W,padx=5)
        self.entauteur=Entry(LeftFrame1, font=('arial',12,'bold'),bd=5,width=44, justify='left',textvariable=auteur)
        self.entauteur.grid(row=2,column=1,sticky=W,padx=5)
        
    
        
        
        self.lbltype=Label(LeftFrame1, font=('arial',12,'bold'),text='Type',bd=7,bg="pink")
        self.lbltype.grid(row=3,column=0,sticky=W,padx=5)
        self.cbotype=ttk.Combobox(LeftFrame1, font=('arial',12,'bold'),width=42,state='readonly',textvariable=type)
        self.cbotype['values']=(' ','Comidique','romantique')
        self.cbotype.current(0)
        self.cbotype.grid(row=3,column=1)
        
            
        self.lblreference=Label(LeftFrame1, font=('arial',12,'bold'),text='Reference',bd=7,bg="pink")
        self.lblreference.grid(row=4,column=0,sticky=W,padx=5)
        self.entreference=Entry(LeftFrame1, font=('arial',12,'bold'),bd=5,width=44, justify='left',textvariable=reference)
        self.entreference.grid(row=4,column=1,sticky=W,padx=5)
        
        
            
        self.lbltitre=Label(LeftFrame1, font=('arial',12,'bold'),text='titre',bd=7,bg="pink")
        self.lbltitre.grid(row=5,column=0,sticky=W,padx=5)
        self.enttitre=Entry(LeftFrame1, font=('arial',12,'bold'),bd=5,width=44, justify='left',textvariable=titre)
        self.enttitre.grid(row=5,column=1,sticky=W,padx=5)
        
        
        #===============================table treeview==============================================
        
        scroll_y=Scrollbar(LeftFrame, orient=VERTICAL)
        self.id_records=ttk.Treeview(LeftFrame,height=14,columns=("idOuvrage","auteur", "type", "reference", "titre"),xscrollcommand=scroll_y.set)
        
        scroll_y.pack(side=RIGHT,fill=Y)
        
        self.id_records.heading("idOuvrage",text="ID")
        self.id_records.heading("auteur",text="Auteur")
        self.id_records.heading("type",text="type")
        self.id_records.heading("reference",text="Reference")
        self.id_records.heading("titre",text="titre")
       #shows heading 
        self.id_records['show']='headings'
        #definir les columns
        self.id_records.column("idOuvrage",width=70)
        self.id_records.column("idOuvrage",width=100)
        self.id_records.column("idOuvrage",width=100)
        self.id_records.column("idOuvrage",width=100)
        self.id_records.column("idOuvrage",width=70)
        
        
        
        self.id_records.pack(fill=BOTH,expand=1)
        #pour eviter d'utiliser le boutton 
        self.id_records.bind("<ButtonRelease-1>",OuvrageInfo)
        #Display()
        
        self.btnAddNew= Button(RightFrame1a,font=('arial',16,'bold'),text="Ajouter",bd=4,pady=1,padx=24,width=8,height=2,command=add,bg='light green').grid(row=0,column=0,padx=1)
        self.btnDisplay= Button(RightFrame1a,font=('arial',16,'bold'),text="Display",bd=4,pady=1,padx=24,width=8,height=2,command=Display,bg='light green').grid(row=1,column=0,padx=1)
        self.btnUpdate= Button(RightFrame1a,font=('arial',16,'bold'),text="Update",bd=4,pady=1,padx=24,width=8,height=2,command=update,bg='light green').grid(row=2,column=0,padx=1)
        self.btnDelete= Button(RightFrame1a,font=('arial',16,'bold'),text="Delete",bd=4,pady=1,padx=24,width=8,height=2,command=delete,bg='light green').grid(row=3,column=0,padx=1)
        self.btnSearch= Button(RightFrame1a,font=('arial',16,'bold'),text="Search",bd=4,pady=1,padx=24,width=8,height=2,command=search,bg='light green').grid(row=4,column=0,padx=1)
        self.btnReset= Button(RightFrame1a,font=('arial',16,'bold'),text="Reset",bd=4,pady=1,padx=24,width=8,height=2,command=Reset,bg='light green').grid(row=5,column=0,padx=1)
        self.btnExit= Button(RightFrame1a,font=('arial',16,'bold'),text="Exit",bd=4,pady=1,padx=24,width=8,height=2,command=iExit,bg='light green').grid(row=6,column=0,padx=1)

        
        
        
        
        
        
if __name__ == "__main__":
        root=Tk()
        application=ouvrage(root)
        root.mainloop()
            
```
