```python
from tkinter import *
from tkinter import ttk
import tkinter.messagebox
import pymysql



class emprunt():
    def __init__(self,root):
        self.root=root
        titlespace=""
        self.root.title(102 * titlespace +"Gestion de bibliothÃ¨que")
        self.root.geometry("800x770+0+0")
        self.root.resizable(width=False, height=False)
        
        MainFrame=Frame(self.root,bd=10,width=800,height=770,relief= RIDGE,bg='light green')
        MainFrame.grid()

        TitleFrame=Frame(MainFrame,bd=10,width=800,height=770,relief=RIDGE)
        TitleFrame.grid(row=0,column=0)
        TopFrame3=Frame(MainFrame,bd=5,width=800,height=770,relief=RIDGE)
        TopFrame3.grid(row=1,column=0)
        
        LeftFrame=Frame(TopFrame3,bd=5,width=800,height=400,padx=2,bg="light green",relief=RIDGE)
        LeftFrame.pack(side=LEFT)
        LeftFrame1=Frame(LeftFrame,bd=5,width=600,height=180,padx=2,pady=4,bg="light green",relief=RIDGE)
        LeftFrame1.pack(side=TOP,padx=0,pady=0)
        
        RightFrame1=Frame(TopFrame3,bd=5,width=100,height=400,padx=2,bg="light green",relief=RIDGE)
        RightFrame1.pack(side=RIGHT)
        RightFrame1a=Frame(RightFrame1,bd=5,width=90,height=300,padx=2,pady=2,bg="light green",relief=RIDGE)
        RightFrame1a.pack(side=TOP)
      #========================Declaration des variabels===============================================================================
        id_emprunt=IntVar()
        date_debut=StringVar()
        date_fin=StringVar()
        #=======================================================================================================
        
        def iExit():
            iExit= tkinter.messagebox.askyesno("Gestion bebliotheque","Confirm if you want to exit")
            if iExit>0:
                root.destroy()
                return
        
        def Reset():
            self.entid.delete(0,END)
            self.entDate_debut.delete(0,END)
            self.entDate_fin.delete(0,END)
            date_fin.set(" ")
            
            
        def add():
            if id_emprunt.get()=="" or date_debut.get()=="" or date_fin.get()=="":
                tkinter.messagebox.showerror("Enter correct details")
            else:
                    sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
                    cur=sqlcon.cursor()
                    cur.execute("insert into emprunt values (%s,%s,%s)",(id_emprunt.get(),date_debut.get(),date_fin.get()))
                    
                    sqlcon.commit()
                    sqlcon.close()
                    
                    tkinter.messagebox.showinfo("Gestion bebliotheque","Record Entered succesfully")
                    
                    
        def Display():
            sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
            cur=sqlcon.cursor()
            cur.execute("select * from emprunt")
            result=cur.fetchall()
            if len(result)!=0:
                self.id_records.delete(*self.id_records.get_children())
                for row in result:
                    self.id_records.insert('',END,values=row)
            sqlcon.commit()
            sqlcon.close()
                    
            tkinter.messagebox.showinfo("Gestion bebliotheque","Record Entered succesfully")
                    
            
            
        def EmpruntInfo(val):
            viewInfo=self.id_records.focus()
            learnerData= self.id_records.item(viewInfo)

            row=learnerData['values']
            id_emprunt.set(row[0])
            date_debut.set(row[1])
            date_fin.set(row[2])
            
        def update():
                    sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
                    cur=sqlcon.cursor()
                    cur.execute("update emprunt set date_debut=%s, date_fin=%s where id_emprunt=%s",(date_debut.get(),date_fin.get()
                                                                                                     ,id_emprunt.get()))
                    
                    sqlcon.commit()
                    sqlcon.close()
                    
                    tkinter.messagebox.showinfo("Gestion bebliotheque","Record updated successfully")
                    
                    
        def delete():
                    sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
                    cur=sqlcon.cursor()
                    cur.execute("delete from emprunt where id_emprunt=%s",id_emprunt.get())
                    
                    sqlcon.commit()
                    Display()
                    sqlcon.close()
                    
                    tkinter.messagebox.showinfo("Gestion bebliotheque","Record deleted successfully")
                    Reset()
        
        def search():
            try:
                sqlcon=pymysql.connect(host='localhost',user='amna',password='',database='Gestion_bibliotheque')
                cur=sqlcon.cursor()
                cur.execute("select * from emprunt where id_emprunt='%s'"%id_emprunt.get())
                row=cur.fetchone()
                    
                id_emprunt.set(row[0])
                date_debut.set(row[1])
                date_fin.set(row[2])
                    
                sqlcon.commit()
                    
            except:  
                tkinter.messagebox.showinfo("Data Entry Form","No such Record Find")
                Reset()
            
            
            sqlcon.close()
                    
    
            
           
            
                    
        #=======================================================================================================
        self.lbltitle=Label(TitleFrame, font=('arial',40,'bold'),text='Gestion_emprunt',bd=7)
        self.lbltitle.grid(row=0,column=0,padx=132)
        
        
        
        
        self.lblid=Label(LeftFrame1, font=('arial',12,'bold'),text='Id',bd=7,bg="light green")
        self.lblid.grid(row=1,column=0,sticky=W,padx=5)
        self.entid=Entry(LeftFrame1, font=('arial',12,'bold'),bd=5,width=44, justify='left',textvariable=id_emprunt)
        self.entid.grid(row=1,column=1,sticky=W,padx=5)
        
        
        
        self.lblDate_debut=Label(LeftFrame1, font=('arial',12,'bold'),text='Date de debut',bd=7,bg="light green")
        self.lblDate_debut.grid(row=2,column=0,sticky=W,padx=5)
        self.entDate_debut=Entry(LeftFrame1, font=('arial',12,'bold'),bd=5,width=44, justify='left',textvariable=date_debut)
        self.entDate_debut.grid(row=2,column=1,sticky=W,padx=5)
        
        
        self.lblDate_fin=Label(LeftFrame1, font=('arial',12,'bold'),text='Date fin',bd=7,bg="light green")
        self.lblDate_fin.grid(row=3,column=0,sticky=W,padx=5)
        self.entDate_fin=Entry(LeftFrame1, font=('arial',12,'bold'),bd=5,width=44, justify='left',textvariable=date_fin)
        self.entDate_fin.grid(row=3,column=1,sticky=W,padx=5)
        
  
        
        
        #===============================table treeview==============================================
        
        scroll_y=Scrollbar(LeftFrame, orient=VERTICAL)
        self.id_records=ttk.Treeview(LeftFrame,height=14,columns=("id_emprunt","date_debut","date_fin"),xscrollcommand=scroll_y.set)
        
        scroll_y.pack(side=RIGHT,fill=Y)
        
        self.id_records.heading("id_emprunt",text="ID")
        self.id_records.heading("date_debut",text="Date debut")
        self.id_records.heading("date_fin",text="Date Fin")
       #shows heading 
        self.id_records['show']='headings'
        #definir les columns
        self.id_records.column("id_emprunt",width=70)
        self.id_records.column("id_emprunt",width=100)
        self.id_records.column("id_emprunt",width=70)
        
        self.id_records.pack(fill=BOTH,expand=1)
        #pour eviter d'utiliser le boutton 
        self.id_records.bind("<ButtonRelease-1>",EmpruntInfo)
        #Display()
        
        
        
        
        
        
        
        
        
        self.btnAddNew= Button(RightFrame1a,font=('arial',16,'bold'),text="Ajouter",bd=4,pady=1,padx=24,width=8,height=2,command=add,bg='pink')
        self.btnAddNew.grid(row=0,column=0,padx=1)
        
        self.btnDisplay= Button(RightFrame1a,font=('arial',16,'bold'),text="Display",bd=4,pady=1,padx=24,width=8,height=2,command=Display,bg='pink')
        self.btnDisplay.grid(row=1,column=0,padx=1)
        
        self.btnUpdate= Button(RightFrame1a,font=('arial',16,'bold'),text="Update",bd=4,pady=1,padx=24,width=8,height=2,command=update,bg='pink')
        self.btnUpdate.grid(row=2,column=0,padx=1)
        
        self.btnDelete= Button(RightFrame1a,font=('arial',16,'bold'),text="Delete",bd=4,pady=1,padx=24,width=8,height=2,command=delete,bg='pink')
        self.btnDelete.grid(row=3,column=0,padx=1)
        
        self.btnSearch= Button(RightFrame1a,font=('arial',16,'bold'),text="Search",bd=4,pady=1,padx=24,width=8,height=2,command=search,bg='pink')
        self.btnSearch.grid(row=4,column=0,padx=1)
        
        self.btnReset= Button(RightFrame1a,font=('arial',16,'bold'),text="Reset",bd=4,pady=1,padx=24,width=8,height=2,command=Reset,bg='pink')
        self.btnReset.grid(row=5,column=0,padx=1)
        
        self.btnExit= Button(RightFrame1a,font=('arial',16,'bold'),text="Exit",bd=4,pady=1,padx=24,width=8,height=2,command=iExit,bg='pink')
        self.btnExit.grid(row=6,column=0,padx=1)


        
        
        
        
        
        
        
        
        
        
if __name__ == "__main__":
        root=Tk()
        application=emprunt(root)
        root.mainloop()
            
```
